```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
import plotly # conda install plotly
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/vcsgithub/vcsgithub.github.io/master/lds41.csv')
df1.head(3)
```

```python
#Line Plot
```

```python
x1 = np.arange(1,6)
y1 = x1 * 2
y2 = x1 ** 2
```

```python
f = plt.figure(figsize=(6,4), dpi=100) # creating figure with custom dimension
plt.plot(x1,y1)
plt.plot(x1,y2)
f = plt.gcf() #get current figure (GCF) default dimension
print(f)
ax = plt.gca() #get current axes - inside element after a canvas
print(ax) # will print only axes as there is only one graph in canvas
print(type(ax))
plt.savefig('dec19plot3.png') #saves the output as impage in path specified
plt.show() # Is the industry standard. Jupiter shows the image by default without this statement
```

```python
print(x1)
print(y1)
print(y2)
plt.plot(x1,y1, color='red', linestyle='dashed', marker='*', label='linear-curve')
plt.plot(x1,y2, color='blue', linestyle='dotted', marker='>', label='expo_curve')
plt.legend() # label will be shown only if this is enabled
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('LinearExpo curve')
#plt.xticks([2,4,6,8,10,12]) # This overrides the default ticks calculated by system
plt.show() # should be called in the end after formatting plot
```

```python
#It is a uni-variet plot, for single value based distribution
np.random.seed(23)
n100 = np.random.randint(low=0, high=101, size=100)
```

```python
plt.hist(x=n100,  bins=10)
```

```python
np.random.seed(23)
pmarks = np.random.randint(low=0, high=101, size=100)
mmarks = np.random.randint(low=0, high=101, size=100)
plt.hist(x=[pmarks, mmarks], bins=10, rwidth=0.90, color=['r','b'])
```

```python
snames=['goog', 'amzn', 'csco', 'infy']
evalues1 = [100,1600,1800,2200]
evalues2 = [1200,1200,200,2400]
```

```python
snames=['goog', 'amzn', 'csco', 'infy']
evalues1 = [100,1600,1800,2200]
evalues2 = [1200,1200,200,2400]

xpos = np.arange(len(snames))
print(xpos)

plt.bar(xpos, evalues1, width=0.4)
plt.bar(xpos+0.4, evalues2, width=0.4)
plt.show()
```
