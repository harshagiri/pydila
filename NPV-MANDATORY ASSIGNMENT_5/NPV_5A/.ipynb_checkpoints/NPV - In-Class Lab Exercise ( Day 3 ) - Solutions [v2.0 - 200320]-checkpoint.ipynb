```python
# create hierarchical index
header = pd.MultiIndex.from_product([['Institute_A','Institute_B'],['Machine Learning','Python']])

# create a list of data
data=([[12,45,67,56],[78,89,45,67],[45,67,89,90],[67,44,56,55]])

# create a dataframe
# index: row labels
df_student = pd.DataFrame(data,
                  index=['Student_1','Student_2','Student_3','Student_4'],
                  columns=header)

# print the dataframe
df_student
```

```python
# stack a dataframe at level 1
# Stacking a dataframe at level 1 will stack the Institute_A and Institute_B column wise
df_stacked_lvl = df_student.stack(level=1)

# print the dataframe
df_stacked_lvl
```
