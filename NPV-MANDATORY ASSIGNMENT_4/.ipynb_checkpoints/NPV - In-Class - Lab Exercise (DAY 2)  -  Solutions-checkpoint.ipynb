```python
# import pandas as pd
import pandas as pd

# create a new dataframe 
df_employee = pd.DataFrame({'Name': ['Emma Larter', 'Mia Junior', 'Sophia Depp', 'James Smith'], 
                   'Salary':[3200, 4500, 3600, 5596]}) 

# print the dataframe   
print("Given Dataframe is :\n",df_employee)
```

```python
# seperate the name column into two seperate columns 
# expand=True: expand the splitted strings into separate columns
df_employee[['First_Name','Last_Name']] = df_employee.Name.str.split(expand=True)

# you can also write the above syntax as:
# df_employee[['First_Name','Last_Name']] = df_employee['Name'].str.split(expand=True)

# print the dataframe
df_employee
```

**Explanation:** By default splitting is done on the basis of single space

**Add the variable 'Salary' as second column**<br>
            
            Salary = [35000, 24000, 55000, 40000]

```python
# store the data
data = {'Name': ['Emma Larter', 'Mia Junior', 'Sophia Depp', 'James Smith'], 'Age': [34, 59, 32, 40]}

# create a dataframe
df_student = pd.DataFrame(data=data)

# print the data
print("Original DataFrame")
print(df_student)

# create a new variable having the elements of the variable 'Salary'
values = [35000, 24000, 55000, 40000]

# insert new variable at index 1
# in python, indexing starts with 0
idx = 1

df_student.insert(loc=idx, column='Salary', value = values)

# print the dataframe
print("\nNew DataFrame")
print(df_student)
```
