```python
# store the data in the form of dictionary
sales = {'Months': ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'],
        'Sales': [22000,27000,25000,29000,35000,67000,78000,67000,56000,56000,89000,60000],
         'Seasons': ['Winter','Winter','Spring','Spring','Spring','Summer','Summer','Summer','Fall','Fall','Fall','Winter']
        }
# create a dataframe
df_sales = pd.DataFrame(sales, columns= ['Months', 'Sales','Seasons'])

# print the dataframe
df_sales
```

```python
# use groupby() method to group the dataframe by 'Seasons'.
# Then use agg() to find minimum, maximum, and average sales for each season
# min: finds the minimum values
# max: finds the maximum values
# mean: calculates the mean
grouped_df = df_sales.groupby('Seasons').agg({'Sales':['min','max','mean']})
# print the dataframe
grouped_df
```
